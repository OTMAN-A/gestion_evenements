   - Affichez tous les résultats avec des messages explicatifs
   - Arrondissez les valeurs à 2 décimales quand approprié

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

print("=== Exercice 2.2 - Analyse statistique ===\n")

# 1. Génération des données
np.random.seed(42)
notes = np.clip(np.random.normal(12, 3, 200), 0, 20)

print("1. Génération des notes:")
print(f"Nombre d'étudiants: {len(notes)}")
print(f"Notes entre {notes.min():.1f} et {notes.max():.1f}\n")

# 2. Statistiques descriptives
print("2. Statistiques descriptives:")
desc = stats.describe(notes)
print(f"Moyenne: {desc.mean:.2f}")
print(f"Médiane: {np.median(notes):.2f}")
print(f"Écart-type: {np.sqrt(desc.variance):.2f}")
print(f"Min: {desc.minmax[0]:.1f}, Max: {desc.minmax[1]:.1f}\n")

# 3. Test de normalité
print("3. Test de normalité:")
stat, p_value = stats.normaltest(notes)
print(f"Statistique: {stat:.3f}")
print(f"P-value: {p_value:.3f}")
if p_value > 0.05:
    print("=> Les notes suivent une loi normale (p > 0.05)")
else:
    print("=> Les notes ne suivent pas une loi normale (p ≤ 0.05)\n")

# 4. Ajustement d'une distribution
print("4. Ajustement d'une loi normale:")
mu, sigma = stats.norm.fit(notes)
print(f"Paramètres: μ = {mu:.2f}, σ = {sigma:.2f}")

# Création du graphique
plt.figure(figsize=(10, 6))
plt.hist(notes, bins=15, density=True, alpha=0.7, color='lightblue', label='Histogramme')

# Courbe de la loi normale ajustée
x = np.linspace(0, 20, 100)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y, 'r-', linewidth=2, label=f'Loi normale (μ={mu:.1f}, σ={sigma:.1f})')

plt.xlabel('Notes')
plt.ylabel('Densité')
plt.title('Distribution des notes et ajustement normal')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 5. Calcul de probabilités
print("5. Calcul de probabilités:")
prob_sup_15 = 1 - stats.norm.cdf(15, mu, sigma)
print(f"Probabilité d'avoir une note > 15: {prob_sup_15:.3f} ({prob_sup_15*100:.1f}%)")

percentile_90 = stats.norm.ppf(0.9, mu, sigma)
print(f"Note du 90ème percentile: {percentile_90:.1f}")
````

## Exercice 1.3 : Optimisation d'un modèle de vente
**Niveau : ⭐⭐ Intermédiaire**

### Énoncé
Une entreprise marocaine de production d'huile d'argan souhaite optimiser son prix de vente.

**Données collectées** (prix en DH, quantité en litres) :
```
Prix :     [150, 180, 200, 220, 250, 280, 300]
Quantité : [850, 780, 720, 650, 580, 500, 420]
```

**Tâches :**

1. Utilisez `curve_fit` pour ajuster un modèle polynomial de degré 2 : `Q(p) = a×p² + b×p + c`

2. Calculez le revenu R(p) = p × Q(p) et trouvez le prix optimal qui maximise le revenu
   - Utilisez `minimize` avec la fonction négative du revenu

3. Visualisez sur un graphique :
   - Les points de données réels
   - La courbe ajustée Q(p)
   - Le point de revenu maximal

4. Affichez le prix optimal et le revenu maximal attendu

### Contraintes
- Le prix doit être entre 100 DH et 400 DH
- Utilisez `scipy.optimize.minimize` avec méthode='L-BFGS-B'


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit, minimize

print("=== Exercice 1.3 - Optimisation du prix de vente ===\n")

# Données de l'entreprise
prix = np.array([150, 180, 200, 220, 250, 280, 300])
quantite = np.array([850, 780, 720, 650, 580, 500, 420])

print("Données de vente:")
print("Prix (DH):", prix)
print("Quantité (L):", quantite)

# 1. Ajustement du modèle polynomial de degré 2
def modele_prix(p, a, b, c):
    return a * p**2 + b * p + c

# Ajuster le modèle aux données
params, _ = curve_fit(modele_prix, prix, quantite)
a, b, c = params

print(f"\nModèle trouvé: Q(p) = {a:.4f}p² + {b:.3f}p + {c:.1f}")

# 2. Calcul du revenu et optimisation
def revenu(p):
    q = a * p**2 + b * p + c
    return -p * q  # négatif car on veut maximiser

# Trouver le prix optimal
resultat = minimize(revenu, 200, method='L-BFGS-B', bounds=[(100, 400)])
prix_optimal = resultat.x[0]
revenu_max = -resultat.fun

print(f"\nPrix optimal: {prix_optimal:.0f} DH")
print(f"Revenu maximal: {revenu_max:.0f} DH")

# 3. Visualisation
p_range = np.linspace(100, 400, 100)
q_range = a * p_range**2 + b * p_range + c
revenu_range = p_range * q_range

plt.figure(figsize=(12, 5))

# Graphique 1: Quantité vs Prix
plt.subplot(1, 2, 1)
plt.scatter(prix, quantite, color='red', s=50, label='Données réelles')
plt.plot(p_range, q_range, 'b-', label='Modèle ajusté')
plt.xlabel('Prix (DH)')
plt.ylabel('Quantité (L)')
plt.title('Quantité vendue vs Prix')
plt.legend()
plt.grid(True, alpha=0.3)

# Graphique 2: Revenu vs Prix
plt.subplot(1, 2, 2)
plt.plot(p_range, revenu_range, 'g-', label='Revenu')
plt.scatter(prix_optimal, revenu_max, color='red', s=100,
           label=f'Optimum: {prix_optimal:.0f} DH')
plt.xlabel('Prix (DH)')
plt.ylabel('Revenu (DH)')
plt.title('Revenu vs Prix')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Quantité vendue au prix optimal
quantite_opt = a * prix_optimal**2 + b * prix_optimal + c
print(f"\nQuantité vendue au prix optimal: {quantite_opt:.0f} L")
````

### Énoncé

Vous analysez la croissance d'une population bactérienne. Les données suivent un modèle exponentiel.

**Tâches :**

1. **Générez des données simulées :**
   ```python
   np.random.seed(42)
   t = np.linspace(0, 10, 20)  # Temps en heures
   # Modèle réel : N(t) = N0 * exp(r*t) avec N0=100, r=0.3
   N_reel = 100 * np.exp(0.3 * t)
   N_observe = N_reel + np.random.randn(20) * 50  # Ajout de bruit
   ```

2. **Définissez le modèle exponentiel :**
   ```python
   def modele_exp(t, N0, r):
       return N0 * np.exp(r * t)
   ```

3. **Ajustez le modèle aux données** avec `curve_fit()`

4. **Visualisez les résultats :**

   - Courbe : modèle ajusté
   - Affichez les paramètres trouvés (N0 et r) dans la légende

5. **Comparez** les paramètres trouvés aux paramètres réels (N0=100, r=0.3)

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

print("=== Exercice 1.4 - Croissance bactérienne ===\n")

# 1. Génération des données simulées
np.random.seed(42)
t = np.linspace(0, 10, 20)  # temps en heures
N_reel = 100 * np.exp(0.3 * t)  # modèle réel
N_observe = N_reel + np.random.randn(20) * 50  # données bruitées

print("Données de croissance bactérienne:")
print(f"Temps: 0 à 10 heures")
print(f"Population initiale réelle: 100 bactéries")
print(f"Taux de croissance réel: 0.3 par heure\n")

# 2. Définition du modèle exponentiel
def modele_exp(t, N0, r):
    return N0 * np.exp(r * t)

# 3. Ajustement du modèle
params, _ = curve_fit(modele_exp, t, N_observe)
N0_trouve, r_trouve = params

print("4. Résultats de l'ajustement:")
print(f"Paramètres trouvés:")
print(f"  N0 = {N0_trouve:.1f} bactéries")
print(f"  r  = {r_trouve:.3f} par heure")
print(f"\nComparaison avec les valeurs réelles:")
print(f"  N0: {N0_trouve:.1f} (trouvé) vs 100.0 (réel)")
print(f"  r:  {r_trouve:.3f} (trouvé) vs 0.300 (réel)")

# 5. Visualisation
plt.figure(figsize=(10, 6))
plt.scatter(t, N_observe, color='red', s=50, label='Données observées', zorder=5)

# Courbe ajustée
t_ajuste = np.linspace(0, 10, 100)
N_ajuste = modele_exp(t_ajuste, N0_trouve, r_trouve)
plt.plot(t_ajuste, N_ajuste, 'b-', linewidth=2,
         label=f'Modèle ajusté: N(t) = {N0_trouve:.0f}·e^{r_trouve:.3f}t')

plt.xlabel('Temps (heures)')
plt.ylabel('Population bactérienne')
plt.title('Croissance bactérienne - Modèle exponentiel')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## Exercice 1.6 : Analyse numérique complète
**Niveau : ⭐⭐⭐ Intermédiaire**

### Énoncé

Vous étudiez le mouvement d'un projectile. Ce problème mobilise plusieurs modules de SciPy.

**Partie A - Interpolation des données expérimentales :**

1. Vous disposez de mesures de la hauteur d'un projectile à différents instants :
   ```python
   t_mesure = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3])
   h_mesure = np.array([0, 11, 18, 21, 20, 15, 6])
   ```

2. Créez une fonction d'interpolation cubique

3. Estimez la hauteur à t = 1.25s et t = 2.25s

**Partie B - Résolution d'équation :**

4. Utilisez `root()` pour trouver l'instant où le projectile retombe au sol (h = 0), en partant de t=2.5

**Partie C - Intégration numérique :**

5. La vitesse verticale est donnée par : v(t) = 20 - 10*t (m/s)

6. Calculez la hauteur maximale atteinte en intégrant v(t) de 0 à 2s avec `quad()`

**Partie D - Visualisation :**

7. Tracez sur un même graphique :
   - Les points de mesure
   - La courbe interpolée
   - Le point d'impact au sol trouvé
   - Une ligne horizontale à la hauteur maximale calculée

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.optimize import root_scalar
from scipy.integrate import quad

print("=== Exercice 1.6 - Mouvement d’un projectile (corrigé) ===\n")

# ---------- données ----------
t_mesure = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3])
h_mesure = np.array([0, 11, 18, 21, 20, 15, 6])

print("Points mesurés :")
for t, h in zip(t_mesure, h_mesure):
    print(f"  t = {t:3.1f} s  →  h = {h:2d} m")

# ---------- A. Interpolation (on autorise l’extrapolation) ----------
interpolation_cubique = interp1d(t_mesure, h_mesure, kind='cubic',
                                fill_value='extrapolate')

h_125 = interpolation_cubique(1.25)
h_275 = interpolation_cubique(2.75)
print(f"\nA. Hauteur estimée :")
print(f"  à t = 1.25 s : {h_125:.1f} m")
print(f"  à t = 2.75 s : {h_275:.1f} m")

# ---------- B. Instant d’impact ----------
def h_func(t):
    return interpolation_cubique(t)

# on cherche où la hauteur s’annule
sol = root_scalar(h_func, bracket=[2.8, 4.0], method='bisect')
t_impact = sol.root
print(f"\nB. Le projectile touche le sol à  t = {t_impact:.2f} s")

# ---------- C. Hauteur max par intégration ----------
def vitesse(t):
    return 20 - 10*t

h_max, _ = quad(vitesse, 0, 2)
print(f"\nC. Hauteur maximale (intégration) : {h_max:.1f} m")

# ---------- D. Visualisation ----------
plt.figure(figsize=(10, 6))
t_fine = np.linspace(0, 4, 300)

plt.scatter(t_mesure, h_mesure, color='red', s=80, zorder=5,
            label='Points mesurés')
plt.plot(t_fine, interpolation_cubique(t_fine),
         'b-', linewidth=2, label='Courbe interpolée')

plt.plot(t_impact, 0, 'ro', markersize=10,
         label=f'Impact  t = {t_impact:.2f} s')
plt.axhline(h_max, color='green', linestyle='--', linewidth=2,
            label=f'Hauteur max = {h_max:.1f} m')

plt.xlabel("Temps (s)")
plt.ylabel("Hauteur (m)")
plt.title("Trajectoire complète du projectile")
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-1, 22)
plt.show()
````

3. Vérifiez la normalité des distributions :
   - Utilisez `shapiro` test
   - Tracez les histogrammes

4. Calculez la corrélation entre avant et après (`pearsonr`)

5. Rédigez une conclusion : le traitement est-il efficace ?


````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

print("=== Chapitre 2 – Exercice 2.1 : Tension artérielle ===\n")

# Données : 30 patients
avant = np.array([145,138,152,148,155,142,149,150,147,151,143,146,154,141,148,
                  153,144,147,150,146,149,152,145,148,151,143,147,150,144,149])
apres = np.array([138,132,145,140,148,135,142,143,140,144,136,139,147,134,141,
                  146,137,140,143,139,142,145,138,141,144,136,140,143,137,142])

print(f"Nombre de patients : {len(avant)}")
print(f"Tension moyenne avant  : {avant.mean():.1f} mmHg")
print(f"Tension moyenne après  : {apres.mean():.1f} mmHg")

# 1. Statistiques descriptives
print("\n1. Statistiques descriptives (via describe) :")
for nom, grp in [("Avant", avant), ("Après", apres)]:
    desc = stats.describe(grp)
    print(f"  {nom:5s} → moyenne={desc.mean:.1f}, médiane={np.median(grp):.1f}, écart-type={np.sqrt(desc.variance):.1f}")

# 2. Test t apparié
t_stat, p_val = stats.ttest_rel(avant, apres)
print(f"\n2. Test t apparié : t = {t_stat:.3f}, p = {p_val:.4f}")
if p_val < 0.05:
    print("   => Différence significative (p < 0.05) : le traitement a un effet.")
else:
    print("   => Pas de différence significative (p ≥ 0.05).")

# 3. Normalité (test de Shapiro-Wilk)
print("\n3. Test de normalité (Shapiro) :")
for nom, grp in [("Avant", avant), ("Après", apres)]:
    w, p = stats.shapiro(grp)
    print(f"  {nom:5s} → W = {w:.3f}, p = {p:.3f} ({'normal' if p > 0.05 else 'non-normal'})")

# 4. Correlation
r, p_r = stats.pearsonr(avant, apres)
print(f"\n4. Correlation (Pearson) : r = {r:.3f}, p = {p_r:.4f}")

# 5. Histogrammes comparatifs
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(avant, bins=8, color='skyblue', edgecolor='black')
plt.title("Avant traitement")
plt.xlabel("Tension (mmHg)")
plt.ylabel("Effectif")

plt.subplot(1, 2, 2)
plt.hist(apres, bins=8, color='lightgreen', edgecolor='black')
plt.title("Après traitement")
plt.xlabel("Tension (mmHg)")
plt.ylabel("Effectif")
plt.tight_layout()
plt.show()

# 6. Conclusion
print("\n5. Conclusion :")
reduction_moy = (avant - apres).mean()
print(f"   Réduction moyenne : {reduction_moy:.1f} mmHg")
if p_val < 0.05:
    print("   Le traitement est efficace sur la tension artérielle.")
else:
    print("   Aucune preuve significative d’efficacité.")
````

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

print("=== Chapitre 2 – Exercice 2.2 : Notes d’examen ===\n")

# 1. Génération des données
np.random.seed(42)
notes = np.clip(np.random.normal(12, 3, 200), 0, 20)

print(f"Nombre d’élèves : {len(notes)}")
print(f"Notes entre {notes.min():.1f} et {notes.max():.1f}\n")

# 2. Statistiques descriptives
desc = stats.describe(notes)
print("2. Statistiques descriptives :")
print(f"  Moyenne    : {desc.mean:.2f}")
print(f"  Médiane    : {np.median(notes):.2f}")
print(f"  Écart-type : {np.sqrt(desc.variance):.2f}")

# 3. Test de normalité (normaltest)
k2, p_norm = stats.normaltest(notes)
print(f"\n3. Test de normalité : K² = {k2:.3f}, p = {p_norm:.3f}")
if p_norm > 0.05:
    print("   → Distribution normale (p > 0.05)")
else:
    print("   → Non normale (p ≤ 0.05)")

# 4. Ajustement d’une loi normale
mu, sigma = stats.norm.fit(notes)
print(f"\n4. Ajustement loi normale : μ = {mu:.2f}, σ = {sigma:.2f}")

# 5. Probabilités
prob_sup_15 = 1 - stats.norm.cdf(15, mu, sigma)
percentile_90 = stats.norm.ppf(0.90, mu, sigma)
print(f"\n5. Probabilités :")
print(f"  P(note > 15)  = {prob_sup_15:.3f} ({prob_sup_15*100:.1f} %)")
print(f"  90ᵉ percentile = {percentile_90:.1f}")

# 6. Graphique : histogramme + PDF ajustée
plt.figure(figsize=(8, 5))
plt.hist(notes, bins=15, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Histogramme')

x = np.linspace(0, 20, 200)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y, 'r-', linewidth=2, label=f'Loi normale\nμ={mu:.1f}, σ={sigma:.1f}')

plt.xlabel("Note")
plt.ylabel("Densité")
plt.title("Distribution des notes et ajustement normal")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

   - Signal original bruité

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import detrend
from scipy.fft import fft, ifft, fftfreq

print("=== Chapitre 3 – Exercice 3.1 : Nettoyage d’un signal audio ===\n")

# 1. Signal bruité (simulation)
np.random.seed(42)
fs = 10_000
t = np.linspace(0, 10, fs*10, endpoint=False)

voix        = np.sin(2*np.pi*150*t) + 0.5*np.sin(2*np.pi*250*t)
bruit_bf    = 0.6*np.sin(2*np.pi*50*t)
bruit_hf    = 0.8*np.sin(2*np.pi*2000*t)
bruit_rand  = 0.3*np.random.randn(len(t))

signal_brute = voix + bruit_bf + bruit_hf + bruit_rand

print("Signal créé :")
print("  – voix 100-300 Hz")
print("  – bruit 50 Hz (basse fréq)")
print("  – bruit 2000 Hz (haute fréq)")
print("  – bruit aléatoire")

# 2. Retrait de la tendance
signal = detrend(signal_brute)

# 3. Analyse spectrale
N = len(signal)
yf = fft(signal)
xf = fftfreq(N, 1/fs)

# 4. Filtrage passe-bande 80-400 Hz
yf_filt = yf.copy()
yf_filt[np.abs(xf) < 80]  = 0
yf_filt[np.abs(xf) > 400] = 0

signal_clean = np.real(ifft(yf_filt))

# 5. Rapport signal/bruit simple
def rms(x):
    return np.sqrt(np.mean(x**2))

snr_before = rms(voix) / rms(signal - voix)
snr_after  = rms(voix) / rms(signal_clean - voix)

print(f"\nRapport signal/bruit (SNR) :")
print(f"  Avant filtrage : {snr_before:.2f}")
print(f"  Après filtrage : {snr_after:.2f}")

# 6. Visualisation 4 sous-graphiques
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# 1. signal brut
axs[0,0].plot(t[:fs], signal_brute[:fs], color='gray', linewidth=0.5)
axs[0,0].set_title("Signal brut (avec bruits)")
axs[0,0].set_xlabel("Temps (s)")
axs[0,0].set_ylabel("Amplitude")

# 2. spectre avant
axs[0,1].plot(xf[:N//2], np.abs(yf[:N//2]), color='red')
axs[0,1].set_title("Spectre avant filtrage")
axs[0,1].set_xlabel("Fréquence (Hz)")
axs[0,1].set_ylabel("|FFT|")
axs[0,1].set_xlim(0, 500)

# 3. spectre après
axs[1,1].plot(xf[:N//2], np.abs(yf_filt[:N//2]), color='green')
axs[1,1].set_title("Spectre après filtrage")
axs[1,1].set_xlabel("Fréquence (Hz)")
axs[1,1].set_ylabel("|FFT|")
axs[1,1].set_xlim(0, 500)

# 4. signal nettoyé
axs[1,0].plot(t[:fs], signal_clean[:fs], color='blue', linewidth=0.5)
axs[1,0].set_title("Signal nettoyé (80-400 Hz)")
axs[1,0].set_xlabel("Temps (s)")
axs[1,0].set_ylabel("Amplitude")

plt.tight_layout()
plt.show()
````

## Exercice 3.2 : Analyse d'un signal avec la FFT
**Niveau : ⭐⭐ Intermédiaire**

### Énoncé

Vous analysez un signal audio composé de plusieurs fréquences mélangées avec du bruit.

**Tâches :**

1. **Création du signal :**
   ```python
   np.random.seed(42)
   fs = 1000  # Fréquence d'échantillonnage (Hz)
   t = np.linspace(0, 1, fs)  # 1 seconde
   
   # Signal composé de 3 sinusoïdes : 50 Hz, 120 Hz, 300 Hz
   signal_pur = 3*np.sin(2*np.pi*50*t) + 1.5*np.sin(2*np.pi*120*t) + 0.8*np.sin(2*np.pi*300*t)
   
   # Ajout de bruit
   bruit = np.random.randn(len(t)) * 2
   signal_bruite = signal_pur + bruit
   ```

2. **Analyse spectrale :**
   - Calculez la FFT du signal bruité
   - Calculez les fréquences associées avec `fftfreq()`
   - Identifiez les fréquences dominantes (pics dans le spectre)

3. **Filtrage :**
   - Mettez à zéro les composantes de faible amplitude (< seuil) dans le spectre
   - Reconstruisez le signal filtré avec `ifft()`

4. **Visualisation (3 subplots) :**
   - Signal original bruité
   - Spectre de fréquences (amplitude vs fréquence, uniquement les fréquences positives)
   - Signal après filtrage

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq

print("=== Chapitre 3 – Exercice 3.2 : Analyse FFT d’un signal ===\n")

# 1. Création du signal
np.random.seed(42)
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

signal_pur = 3*np.sin(2*np.pi*50*t) + 1.5*np.sin(2*np.pi*120*t) + 0.5*np.sin(2*np.pi*300*t)
bruit = 2*np.random.randn(len(t))
signal_bruite = signal_pur + bruit

print("Signal créé :")
print("  Fréquences pures : 50 Hz, 120 Hz, 300 Hz")
print("  Bruit gaussien ajouté")

# 2. FFT
N = len(signal_bruite)
yf = fft(signal_bruite)
xf = fftfreq(N, 1/fs)

# 3. Identification des pics (fréquences dominantes)
amplitudes = np.abs(yf[:N//2])
pics_idx = np.where(amplitudes > 30)[0]  # seuil simple
pics_freq = xf[pics_idx]

print(f"\n3. Fréquences dominantes détectées : {pics_freq} Hz")

# 4. Filtrage – on garde seulement les pics
yf_filt = np.zeros_like(yf)
seuil = 30
for i in range(N//2):
    if amplitudes[i] > seuil:
        yf_filt[i] = yf[i]
        yf_filt[N-i-1] = yf[N-i-1]  # symétrie

signal_filtre = np.real(ifft(yf_filt))

# 5. Graphiques
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(t[:200], signal_bruite[:200], color='gray', linewidth=0.8)
plt.title("Signal bruité")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")

plt.subplot(1, 3, 2)
plt.stem(xf[:N//2], amplitudes, basefmt=" ")
plt.title("Spectre (amplitudes)")
plt.xlabel("Fréquence (Hz)")
plt.ylabel("|FFT|")
plt.xlim(0, 400)

plt.subplot(1, 3, 3)
plt.plot(t[:200], signal_filtre[:200], color='blue', linewidth=0.8)
plt.title("Signal filtré")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
````

## Exercice 3.3 : Détection de contours dans une image médicale
**Niveau : ⭐⭐⭐ Avancé**

### Énoncé
Vous développez un outil d'aide au diagnostic pour détecter des anomalies sur des radiographies pulmonaires.

**Tâches :**

2. Prétraitement de l'image :
   - Appliquez un filtre gaussien (`ndimage.gaussian_filter`)
   - Testez différentes valeurs de sigma (1, 2, 5)

3. Détection de contours :
   - Calculez le gradient avec `ndimage.sobel`
   - Appliquez un seuillage pour isoler les contours nets
   - Utilisez `ndimage.binary_closing` pour combler les gaps

4. Segmentation des régions :
   - Utilisez `ndimage.label` pour identifier les zones connexes
   - Comptez le nombre de structures détectées
   - Calculez l'aire de chaque structure

   - Image originale

   - Contours détectés
   - Régions segmentées (avec labels colorés)

**Bonus :** Calculez le centre de masse de chaque structure avec `ndimage.center_of_mass`


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter, sobel, binary_closing, label, center_of_mass

print("=== Chapitre 3 – Exercice 3.3 : Détection de contours (image médicale) ===\n")

# 1. Image synthétique 256×256
np.random.seed(42)
image = np.full((256, 256), 100, dtype=np.float32)  # fond gris

# 3 cercles = poumons + anomalie
def cercle(img, cx, cy, r, val):
    y, x = np.ogrid[:img.shape[0], :img.shape[1]]
    mask = (x - cx)**2 + (y - cy)**2 <= r**2
    img[mask] = val
    return img

cercle(image, 128, 100, 45, 180)   # poumon gauche
cercle(image, 128, 156, 40, 180)   # poumon droit
cercle(image,  90, 128, 15, 220)   # petite anomalie

image += np.random.normal(0, 8, image.shape)  # bruit gaussien

print("Image 256×256 créée : fond 100, poumons 180, anomalie 220 + bruit")

# 2. Pré-traitement : filtre gaussien
image_filt = gaussian_filter(image, sigma=2)

# 3. Détection contours : gradient Sobel + seuil
grad = np.hypot(sobel(image_filt, axis=0), sobel(image_filt, axis=1))
seuil = 30
contours = grad > seuil
contours = binary_closing(contours, iterations=1)  # ferme les trous

# 4. Segmentation : labellisation
label_image, nb_obj = label(contours)
print(f"\n4. Nombre de structures détectées : {nb_obj}")

# 5. Aires et centres de masse
aires = [np.sum(label_image == i) for i in range(1, nb_obj + 1)]
centres = center_of_mass(image, label_image, range(1, nb_obj + 1))

print("  Aires (pixels) :", [int(a) for a in aires])
print("  Centres (ligne, colonne) :", [tuple(np.round(c, 1)) for c in centres])

# 6. Visualisation complète
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].imshow(image, cmap='gray')
axes[0, 0].set_title("Image originale")

axes[0, 1].imshow(image_filt, cmap='gray')
axes[0, 1].set_title("Après filtre gaussien")

axes[1, 0].imshow(contours, cmap='gray')
axes[1, 0].set_title("Contours détectés")

# image colorée par label
axes[1, 1].imshow(label_image, cmap='prism', interpolation='nearest')
axes[1, 1].set_title("Régions segmentées")

for ax in axes.flat:
    ax.axis('off')
plt.tight_layout()
plt.show()
````

````python
# ========== Partie 1 : Chargement & exploration ==========
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns

# 1-1 Chargement
iris = pd.read_csv('Iris.csv')          # ou '/content/Iris.csv' si besoin
print("1-1 Chargement OK")
print(iris.head())
print("Colonues :", iris.columns.tolist())

# 1-2 Statistiques descriptives (scipy.stats.describe) pour chaque colonne numérique
print("\n1-2 Describe par variable (scipy.stats.describe) :")
for col in iris.select_dtypes(include=np.number).columns:
    d = st.describe(iris[col])
    print(f"{col:15s} : moy={d.mean:5.2f}  var={d.variance:6.2f}  min={d.minmax[0]:5.1f}  max={d.minmax[1]:5.1f}")


# 1-3 Test de normalité (Shapiro) pour chaque colonne numérique
print("\n1-3 Test de normalité (Shapiro) :")
for col in iris.select_dtypes(include=np.number).columns:
    stat, p = st.shapiro(iris[col])
    print(f"{col:15s} : W={stat:.3f}, p={p:.3f} {'normal' if p>0.05 else 'non-normal'}")

# 1-4 Matrice de corrélation
corr = iris.select_dtypes(include=np.number).corr(method='pearson')
print("\n1-4 Matrice de corrélation :")
print(corr.round(3))

# carte de chaleur rapide
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matrice de corrélation - Iris")
plt.show()
````

````python
# ========== Partie 2 : KDTree + KNN manuel ==========
from scipy.spatial import KDTree

# on garde seulement les 4 colonnes numériques
X = iris.select_dtypes(include=np.number).values
y = iris['Species'].values          # ou iris.iloc[:, -1] selon ton csv

# 80 % train / 20 % test
np.random.seed(42)
indices = np.random.permutation(len(X))
train_idx = indices[:int(0.8*len(X))]
test_idx  = indices[int(0.8*len(X)):]

X_train, y_train = X[train_idx], y[train_idx]
X_test,  y_test  = X[test_idx],  y[test_idx]

print("2-1 Split 80/20 :")
print(f"  Train : {len(X_train)} échantillons")
print(f"  Test  : {len(X_test)} échantillons")

# construction KDTree
kdt = KDTree(X_train)

# 2-2 & 2-3 : prédiction k=5 + vote majoritaire
k = 5
preds = []

for x in X_test:
    dist, idx = kdt.query(x, k=k)
    voisins_classe = y_train[idx]
    # vote majoritaire
    unique, counts = np.unique(voisins_classe, return_counts=True)
    preds.append(unique[counts.argmax()])

# 2-4 accuracy
accuracy = np.mean(np.array(preds) == y_test)
print(f"\n2-4 Accuracy (k={k}) : {accuracy:.3f}")

# 2-5 affichage des premières prédictions
print("\n2-5 Exemples de prédictions (5 premiers) :")
for i in range(5):
    print(f"  Vrai : {y_test[i]:15s}  |  Pred : {preds[i]:15s}")
````

````python
# ========== Partie 3 : Optimisation de k ==========
k_values = range(1, 21)
acc_list = []

for k in k_values:
    preds = []
    for x in X_test:
        dist, idx = kdt.query(x, k=k)
        voisins_classe = y_train[idx]
        unique, counts = np.unique(voisins_classe, return_counts=True)
        preds.append(unique[counts.argmax()])
    acc = np.mean(np.array(preds) == y_test)
    acc_list.append(acc)

# 3-1 Meilleur k
best_k = k_values[np.argmax(acc_list)]
best_acc = max(acc_list)
print("3-1 Accuracy vs k :")
for k, acc in zip(k_values, acc_list):
    print(f"  k={k:2d} → accuracy={acc:.3f}")
print(f"\nMeilleur k : {best_k}  (accuracy = {best_acc:.3f})")

# 3-2 Courbe
plt.figure(figsize=(7, 4))
plt.plot(k_values, acc_list, marker='o', color='teal')
plt.title("Accuracy en fonction de k")
plt.xlabel("k (nombre de voisins)")
plt.ylabel("Accuracy")
plt.xticks(k_values)
plt.grid(True, alpha=0.3)
plt.axvline(best_k, color='red', linestyle='--', label=f'k optimal = {best_k}')
plt.legend()
plt.show()
````

````python
# ========== Partie 4 : Matrice de confusion & métriques ==========
from scipy.stats import contingency

# on reprend le meilleur k trouvé dans la partie 3
k_opt = best_k
preds_opt = []
for x in X_test:
    dist, idx = kdt.query(x, k=k_opt)
    voisins_classe = y_train[idx]
    unique, counts = np.unique(voisins_classe, return_counts=True)
    preds_opt.append(unique[counts.argmax()])

# 4-1 Matrice de confusion (sans scipy.stats, 100 % manuel)
import pandas as pd

# DataFrame croisé
conf_df = pd.crosstab(y_test, preds_opt, rownames=['Vrai'], colnames=['Pred'])
mat_conf = conf_df.values
print("4-1 Matrice de confusion (lignes = vrai, colonnes = pred) :")
print(conf_df)

# 4-2 Métriques par classe
classes = np.unique(y_test)
precision = {}
recall    = {}
f1        = {}

for i, cls in enumerate(classes):
    tp = mat_conf[i, i]
    fp = mat_conf[:, i].sum() - tp
    fn = mat_conf[i, :].sum() - tp

    precision[cls] = tp / (tp + fp) if (tp + fp) else 0
    recall[cls]    = tp / (tp + fn) if (tp + fn) else 0
    f1[cls]        = 2 * precision[cls] * recall[cls] / (precision[cls] + recall[cls]) if (precision[cls] + recall[cls]) else 0

print("\n4-2 Métriques par classe :")
for cls in classes:
    print(f"  {cls:17s} : precision={precision[cls]:.3f}, recall={recall[cls]:.3f}, F1={f1[cls]:.3f}")

# 4-3 Barplot des F1
plt.figure(figsize=(6, 4))
plt.bar(classes, [f1[c] for c in classes], color=['skyblue', 'lightgreen', 'salmon'])
plt.title(f"F1-score par classe (k={k_opt})")
plt.ylabel("F1-score")
plt.ylim(0, 1.05)
for i, c in enumerate(classes):
    plt.text(i, f1[c] + 0.02, f"{f1[c]:.3f}", ha='center')
plt.tight_layout()
plt.show()
````

### Énoncé

Vous travaillez sur un projet de maintenance prédictive. Vous disposez de données de vibration d'une machine industrielle. Vous devez analyser ces signaux pour détecter des anomalies.

**Données simulées :**
```python
np.random.seed(42)

# Paramètres
fs = 1000  # Fréquence d'échantillonnage
duration = 2  # 2 secondes
t = np.linspace(0, duration, fs * duration)

# Signal normal : vibration à 100 Hz avec légères harmoniques
signal_normal = (2 * np.sin(2*np.pi*100*t) +
                 0.5 * np.sin(2*np.pi*200*t) +
                 np.random.randn(len(t)) * 0.3)

# Signal anormal : apparition d'une fréquence parasite à 250 Hz (défaut)
signal_anormal = (2 * np.sin(2*np.pi*100*t) +
                  0.5 * np.sin(2*np.pi*200*t) +
                  1.5 * np.sin(2*np.pi*250*t) +  # Fréquence de défaut
                  np.random.randn(len(t)) * 0.3)
```

**Tâches :**

1. **Prétraitement :**
   - Supprimez la tendance des deux signaux avec `signal.detrend()`

2. **Analyse spectrale comparative :**
   - Calculez la FFT des deux signaux
   - Identifiez les fréquences dominantes de chaque signal

3. **Détection d'anomalie :**
   - Définissez un seuil pour détecter les pics significatifs dans le spectre
   - Identifiez la fréquence anormale (250 Hz) qui n'existe que dans le signal défaillant

4. **Statistiques descriptives :**
   - Utilisez `stats.describe()` pour comparer les deux signaux
   - Comparez les variances et les valeurs extrêmes

5. **Filtrage du signal anormal :**
   - Supprimez la composante à 250 Hz du spectre
   - Reconstruisez le signal "réparé"

6. **Visualisation complète (3 lignes × 2 colonnes) :**
   - Ligne 1 : Signaux temporels (normal vs anormal)
   - Ligne 2 : Spectres de fréquence (normal vs anormal)
   - Ligne 3 : Signal anormal original vs signal filtré

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import detrend
from scipy.fft import fft, ifft, fftfreq
from scipy import stats

print("=== Exercice 3.6 (conforme PDF) – Analyse vibration IoT ===\n")

# ---------- 1. DONNÉES SIMULÉES ----------
fs = 1000
duration = 2
t = np.linspace(0, duration, fs*duration, endpoint=False)

signal_normal = (2*np.sin(2*np.pi*100*t) +
                 0.5*np.sin(2*np.pi*200*t) +
                 0.3*np.random.randn(len(t)))

signal_anomal = (2*np.sin(2*np.pi*100*t) +
                 0.5*np.sin(2*np.pi*200*t) +
                 1.5*np.sin(2*np.pi*250*t) +   # défaut
                 0.3*np.random.randn(len(t)))

# ---------- 2. RETRAIT TENDANCE ----------
sig_n = detrend(signal_normal)
sig_a = detrend(signal_anomal)

# ---------- 3. FFT ----------
N = len(sig_n)
yf_n = fft(sig_n)
yf_a = fft(sig_a)
xf = fftfreq(N, 1/fs)

# ---------- 4. DÉTECTION PICS ----------
def pics_significants(yf, xf, seuil=25):
    amp = np.abs(yf[:N//2])
    return xf[np.where(amp > seuil)]

pics_n = pics_significants(yf_n, xf)
pics_a = pics_significants(yf_a, xf)
print("Fréquences dominantes :")
print("  Normal :", [f"{f:.0f} Hz" for f in pics_n])
print("  Anomal :", [f"{f:.0f} Hz" for f in pics_a])

# ---------- 5. FILTRAGE 250 Hz ----------
yf_filt = yf_a.copy()
yf_filt[np.abs(xf - 250) < 0.5] = 0
signal_repare = np.real(ifft(yf_filt))

# ---------- 6. STATISTIQUES DESCRIPTIVES ----------
desc_n = stats.describe(sig_n)
desc_a = stats.describe(sig_a)
print("\nStatistiques (après dé-trend) :")
print("  Normal : moy=%.3f  var=%.3f  min=%.3f  max=%.3f" %
      (desc_n.mean, desc_n.variance, desc_n.minmax[0], desc_n.minmax[1]))
print("  Anomal : moy=%.3f  var=%.3f  min=%.3f  max=%.3f" %
      (desc_a.mean, desc_a.variance, desc_a.minmax[0], desc_a.minmax[1]))

# ---------- 7. VISUAL 3×2 ----------
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

# ligne 1 : temporel
axes[0,0].plot(t[:500], sig_n[:500], color='b'); axes[0,0].set_title("Normal détrend")
axes[0,1].plot(t[:500], sig_a[:500], color='r'); axes[0,1].set_title("Anomal détrend")

# ligne 2 : spectre
axes[1,0].stem(xf[:N//2], np.abs(yf_n[:N//2]), basefmt=" "); axes[1,0].set_title("Spectre normal")
axes[1,1].stem(xf[:N//2], np.abs(yf_a[:N//2]), basefmt=" "); axes[1,1].set_title("Spectre anomal")
for ax in axes[1,:]: ax.set_xlim(0, 400)

# ligne 3 : réparation
axes[2,0].plot(t[:500], sig_a[:500], color='r'); axes[2,0].set_title("Avant filtrage")
axes[2,1].plot(t[:500], signal_repare[:500], color='g'); axes[2,1].set_title("Après suppression 250 Hz")
for ax in axes[2,:]: ax.set(xlabel='Temps (s)', ylabel='Amplitude')

plt.tight_layout(); plt.show()
````
